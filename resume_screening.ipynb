## Sample Text Data

```python
resume = """
Python developer with skills in Machine Learning,
Deep Learning, TensorFlow, CNN and data analysis.
Experienced with Git and GitHub.
"""

job_description = """
Looking for an AI/ML intern with knowledge of Python,
Machine Learning, Deep Learning, TensorFlow and version control.
"""
```

## Creating Processing Function

```python
def preprocess_text(text):
    
    # Covert to lowercase 
    text = text.lower()

    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenization
    words = word_tokenize(text)

    # Load stopwords
    stop_words = set(stopwords.words('english'))

    # Remove stopwords
    filtered_words = [
        word for word in words
        if word not in stop_words
    ]

    # Join words back
    cleaned_text = ' '.join(filtered_words)

    return cleaned_text
```

### Call the Function

```python
cleaned_resume = preprocess_text(resume)
cleaned_job = preprocess_text(job_description)

print("Cleaned Resume:\n")
print(cleaned_resume)

print("\nCleaned Job Description:\n")
print(cleaned_job)
```

### Import TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

### Creating TF-IDF Object

```python
vectorizer = TfidfVectorizer()
```

### Creating TF-IDF Vectors

```python
documents = [cleaned_resume, cleaned_job]
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
# View vocabulary
print(vectorizer.get_feature_names_out())
```

```python
print(tfidf_matrix.toarray())
```

### Importing Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

### Calculating Similarity

```python
similarity_score = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print(similarity_score)
```

```python
match_percentage = similarity_score[0][0]*100
print(f"Resume Match Score: {match_percentage:.2f}%")
```

```python
import PyPDF2
```

#### PDF Extraction System

```python
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        
        for page in reader.pages:
            text += page.extract_text()

    return text
```

```python
resume_text = extract_text_from_pdf("D:\Deep Learning\Projects\Multi Task Models\data\Final-Resume.pdf")

print(resume_text)
```

### Complete Pipeline

```python
# Extract text from PDF
resume_text = extract_text_from_pdf("D:\Deep Learning\Projects\Multi Task Models\data\Final-Resume.pdf")

# Preprocess extracte3d resume text
cleaned_resume = preprocess_text(resume_text)

# Preprocess Job description 
cleaned_job = preprocess_text(job_description)

# Create document list
documents = [cleaned_resume, cleaned_job]

# Convert text into TF-IDF vectors 
tfidf_matrix = vectorizer.fit_transform(documents)

# Calculate cosine similarity
similarity_score = cosine_similarity(
    tfidf_matrix[0:1],
    tfidf_matrix[1:2]
)

# Convert score into percentage
match_percentage = similarity_score[0][0]*100

# Print final result
print(f"Resume Match Score: {match_percentage:.2f}%")
```
